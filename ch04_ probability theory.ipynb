
# 4장 확률이론

## 1. 확률의 정의

- 상대빈도정의 : 실험을 무한에 가깝게 시행했을 때, 전체 시행횟수에서 그 사건이 나타나는 빈도수를 상대적으로 나타낸 것
    - $P(A) = lim_{N →\infty} \frac{n}{N}$
        - N = 총 시행 횟수, n = A사건이 발생한 횟수
- 동등발생정의 : 모든 경우들이 각각 동일한 가능성을 가지고 발생할 것 이라는 가정하에서 특정 사건이 일어날 확률을 정의한 것
    - $P(A) = \frac{사건 A에 속하는 경우의 수}{발생할 가능성이 동일한 전체 경우의 수}$

## 2. 집합이론과 확률의 개념

- 집합이론
    - 집합 : 개체 또는 원소의 모임.
    - 원소 : 집합을 형성하는 개별사물
        - A = 집합 = {남자,여자} ={집합원소}
    - 전체집합 : 특정한 문제에서 가능한 모든 원소의 집합
    - 부분집합 : 모든 원소 중 일부를 나타내는 집합
- 집합연산
    - 여집합$(\bar{A})$ : 전체 집합 중 집합 A에 포함되지 않는 원소들
    - 합집합($A \cup B$) : 집합 A또는 집합 B에 속하는 원소
    - 교집합($A \cap B$) : 집합A와 B의 공통원소
    - 합집합의 계산 : $A \cup B = A + B - A \cap B$
    - 배타적집합 : $A \cap B = \emptyset$
- 집합이론과 확률이론
    - 실험이나 관찰에서 얻은 결과 = 사건 or 사상 (=집합)
    - 여러 개 사건 중 한 사건 = 단일사건 또는 경우 (=원소)
    - 실험이나 관찰에서 생길 수 있는 모든 사건의 모임 = 표본공간 (전체집합)

## 3. 기본적인 확률법칙

- 확률의 덧셈법칙 : 집합이론에서 합집합의 계산에 대응됨
    - $P(A \cup U) = P(A) + P(B) - P(A \cap B)$
    - 배타적 사건일 때 : $P(A \cup B) = P(A) + P(B)$
    - 배타적 사건의 예
        - P(90점 이상) = 10/60, P(80점 미만) = 30/60
        - 학생의 점수가 90점 이상이면서 80점 미만일 확률? 없음 → 배타적
        - 주사위를 던질 때 점이 1이면서 2가 나올 확률? 없음 → 베타적
    - 서로 배타적이 아닌 예
        - 60명의 필기구 조사 연필 가진 학생(A) = 30명, 볼펜 가진 학생(B) = 40명
        - $A \cap B = 25$
        - 학생을 임의로 한명 뽑을 때 볼펜 또는 연필을 가진 학생이 뽑힐 확률
            - $P(A \cup B)$ $= P(A) + P(B) - P(A \cap B)$ = 30/60 + 40/60 -25/60
                
                                    = 45/60
                
- 조건부확률
    
    : 바로 전의 실험결과에 의해 변화된 표본공간에서 어떤 사건이 일어날 확률
    
    - $P(B | A) = \frac{P(A \cap B)}{P(A)}$
- 확률의 곱셈법칙 : 집합이론에서 교집합의 개념에 대응됨
    - $P(A \cap B) = P(B) \cdot P(A|B) = P(A) \cdot P(B|A)$
    - 예시 : 총 학생수 = 10,000, 4학년 = 2,000, 여학생=3,500, 여학생 중 4학년=8/35
    - 무작위 선택 시 4학년 여학생일 확률
        - $P(여학생 \cap 4학년) = P(여학생) \cdot P(4학년|여학생)$ = 35/10 x 8/35 = 0.08
- 곱셈법칙과 의사결정수
    - 의사결정수 : 여러단계를 거치는 확률실험에서 나타날 수 있는 모든 가능한 결과
- 독립사건과 종속사건
    - 독립사건 : 처음에 어떤 결과가 나왔느냐 하는 것이 다음 사건에 영향을 주지 않음
        - $P(A|B) = P(A), P(B|A) = P(B)$
        - 곱셈법칙 : $P(A \cap B) = P(A) \cdot P(B)$
    - 종속사건 : 한 사건의 발생이 다음에 발생할 사건에 영향을 주는 경우 = 조건부확률
        - 곱셈법칙 : $P(A \cap B) = P(B) \cdot P(A|B) = P(A) \cdot P(B|A)$ 기존과 동일
- 복원추출과 비복원추출
    - 복원추출 : 표본을 뽑은 후 다시 복원 시킴 (독립적=표본공간의 변화X)
        - ex. 처음 꺼낸 공을 다시 집어넣지 않음
    - 비복원추출 : 표본을 뽑은 후 다시 집어넣지 않음 (종속적)
        - ex. 처음 꺼낸 공을 다시 집어 넣음

## 4. 베이즈정리

: 사전에 알고 있는 정보에 기준을 두고 어떤 사건이 일어나게 될 확률을 계산하는 이론

- $P(A|B) = \frac{P(A \cap B)}{P(B)}$ or $P(A_k|B) = \frac{P(A_k \cap B)}{\sum P(A_i \cap B)} = \frac{P(A_k) \cdot P(B|A_k)}{\sum P(A_i) \cdot P(B|A_i)}$


# 8장 표본 및 표집분포

## 1. 표본조사와 표본추출

- 표본조사의 필요성
    - 경제성, 시간의 제약, 무한모집단, 조사 불가 모집단, 정확성, 그밖의 이유
- 표본추출방법
    - 표본조사에 의한 모집단의 이해는 필연적으로 오차를 수반함
    - 표본추출의 오차는 편의(bias)와 우연성(chance)에 의해 발생
        - 우연에 의한 오차 : 표본 크기 증가로 감소 가능
        - 편의에 의한 오차 : 표본추출방법을 과학적으로 계획함으로 줄일 수 있음
- 확률표본추출(=무작위 추출, random sampling)
    - 각 구성원이 표본으로 선택될 가능성을 일정하게 하는 표본추출방법
    - 편의는 제거되지만 우연성 오류는 없어지지 않음 → 정규분포를 이뤄서 예측 가능
    - 가장 많이 쓰이는 방법
        - 단순무작위추출(simple random sampling)
            - 모집단에 포함되어 있는 모든 구성원이 뽑힐 확률을 각각 같도록 하는 추출방법
        - 층별추출(stratified sampling)
            - 모집단을 성격에 따라 여러 집단 또는 여러층으로 분류한 후 각 층에서 표본을 추출
        - 군집추출(cluster sampling)
            - 자연적 또는 인위적으로 집단을 먼저 뽑고, 그 집단 중 필요한 만큼의 표본을 추출
        - 체계적추출(systematic sampling)
            - 하나의 모집단배열이 무작위로 되어있을 때 체계적 수단을 동원하여 추출하는 방법
- 비확률표본추출
    - 확률추출이 불가능하거나 비경제적일 경우 사용. 연구자가 모집단과 비슷하다고 생각되는 표본을 임의로 추출해 내는 방법. 주관에 의해 선택됨으로 오차에 대한 분석이 불가능.
        - 편의추출(convenience sampling)
            - 연구자가 가장 손쉽게 구할 수 있는 구성원을 선택하여 표본으로 삼음(ex. 모교)
        - 판단추출(judgment sampling)
            - 전문지식이 있는 사람이 판단하기에 가장 효과적이라 생각하는 표본(ex. 물가지수)

## 2. 표본추출오차와 비표본추출오차

- 표본추출오차(sampling error)
    - 모집단을 대표할 수 있는 전형적인 구성요소를 표본으로 선택하지 못했기 때문에 발생하는 오류로 두 가지 요인이 있음
        - 표본의 크기 때문에 생기는 우연에 의한 오류
        - 모집단을 대표할 수 없는 비전형적인 구성요소를 표본으로 뽑았기 때문에 일어나는 오류
- 비표본추출오차(non-sampling error)
    - 표본선택이 잘못된 것이 아니라 주로 표본의 특성값을 측정하는 방법이 부정확하기 때문
        - 측정오차(measurement error)

## 3. 통계량과 표집분포

- 모수 : 모집단의 분포 특성
    - 유한모집단(finite population) : 모집단의 구성원 수가 유한
        - 모수 계산
            - 평균 : $\mu = \frac{\sum X_i}{N}$
            - 분산 : $\sigma^2 = \frac{\sum (X_i - \mu)^2}{N}$
            - 표준편차 : $\sigma = \sqrt{\sigma^2}= \sqrt{ \frac{\sum (X_i - \mu)^2}{N}}$
    - 무한모집단(infinite population) : 모집단의 구성원 수가 무한
        - 모수 계산
            - 평균 : $\mu = E(X) = \sum P_i X_i$
            - 분산 : $\sigma^2 = \sigma^2(X) =  \sum [X - E(X)]^2 \cdot P(X_i)$
            - 표준편차 : $\sigma = \sqrt{\sigma^2} = \sqrt{  \sum [X - E(X)]^2 \cdot P(X_i) }$
- 통계량(statistic) : 표본의 분포 특성
    - 표본의 분포특성을 계수화하기 위해 사용되는 방법
    - 표본의 분산이나 표준편차 계산시에는 n대신 n-1을 사용
    - 통계량 계산
        - 평균 : $\bar{X} = \frac{\sum X_i}{n}$
        - 분산 : $S^2 = \frac{\sum (X_i - \bar{X})^2}{n-1}$
        - 표준편차 : $S = \sqrt{S^2} = \sqrt{\frac{\sum (X_i - \bar{X})^2}{n-1}}$표
- 표집분포(sampling distribution)
    - 모집단에서 일정한 크기로 뽑을 수 있는 표본을 모두 뽑았을 때, 그 모든 표본의 특성치, 즉 통계량의 확률분포를 말함

## 4.평균의 표집분포의 평평균의

- 평균의 표집분포
    
    : 특정한 모집단에서 동일한 크기로 가능한 모든 표본을 뽑아서 각각평균이
    
    - 평균의 표집분포의 평균
        - $\mu_{\bar{X}} = \sum \bar{X}_i \cdot P(\bar{X}_i)$
    - 평균의 표집분포의 평균과 모집단 평균
        - $\mu_{\bar{X}} = \mu$
    - 평균의 표집분포의 분산
        - $\sigma^{2}_{\bar{X}} = E(\bar{X}- \mu_{\bar{X}})^2 =  \sum (\bar{X}- \mu_{\bar{X}})^2 \cdot P(\bar{X})$펴
    - 평균의 표집분포의 분산과 표준편차
        - 분산
            - $\sigma^{2}_{\bar{X}} = \frac{\sigma^{2}}{n}$
        - 표준편차
            - $\sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}}$
- 모집단의 분포와 평균의 표집분포
    - 모집단의 정규분포 여부와 표본크기 n이 변화함에 따라 달라지는 표집분포의 모양
    - 모집단이 정규분포일 때 평균의 표집분포
        - $\mu_{\bar{X}} = \mu$
        - $\sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}}$
        - $\bar{X}$에 해당되는 Z값  $= Z = \frac{\bar{X} - \mu_{\bar{X}}}{\sigma_{\bar{X}}} = \frac{ \bar{X}-\mu_{\bar{X}}}{\sigma /{\sqrt{n}}}$
- 모집단이 정규분포가 아닐 때
    - 중심극한정리
        
        : 모집단의 분포모양과는 관계없이 평균과 분산이 있는 모집단에서 크기가 $n$인 선택가능한 모든 표본을 뽑을 때 평균의 표집분포는 $n$을 증가 시킬 수록 정규분포에 접근하게 된다는 것
        
        (보통 n=30 이상이면 평균의 표집분포는 정규분포를 이룸)
        
        - $\mu_{\bar{X}} = \mu$
        - $\sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}}$
- 모집단의 크기가 작을 때
    - 모집단의 크기가 작더라도 복원추출시 무한모집단 같은 효과. 비복원추출의 경우 조정 필요
    - 모집단이 작을 때 표본평균($\bar{X}$)의 표준편차
        - $\sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}} \cdot \sqrt{\frac{N-n}{N-1}}$
            - $N$ : 모집단의 크기, $n$ : 표본의 크기
            - $\sqrt{{N-n}/{N-1}}$ : $\bar{X}$-분포의 표준편차 계산을 위한 조정계수
                - 표본의 크기가 작을 때는 조정계수가 1에 가까워짐
                - 표본의 크기가 커질 수록 조정계수의 값이 1보다 작아짐

## 5. 분산 및 비율의 표집분포

- 분산의 표집분포
    
    : 모집단이  $\sigma^2$의 분산을 가질 때 모집단으로부터 크기가 동일하게 선택 가능한 모든 표본을 뽑아서 각각의 분산을 계산했을 때, 표본 분산 $S^2$들은 일정한 분포를 이루는 것
    
    - 분산의 표집분포의 평균과 모집단의 분산
        - $E(S^2) = \sigma^2$
    - $\chi^2$ - 분포
        
        : 모집단의 분포가 정규분포인 경우 위와 같은 수식이고 분산의 표집분포 모양은 $n$의 크기에 따라 해당 분포를 이룬다.
        
        - 비대칭의 모양을 이루고, 오른쪽으로 긴 꼬리를 가지며, 항상 양수값만을 갖는 특징
        - $\chi^{2}_{n-1} = \frac{(n-1) \cdot S^2}{\sigma^2}$
            - $n-1$ = 자유도라고 하며, 분포의 모양이 달라짐. 커질수록 정규분포에 가깝게 됨
- 비율의 표집분포(sampling distribution of proportion)
    
    : 동일한 모집단에서 선택 가능한 모든 표본을 뽑아 구한 비율들의 분포
    
    - 전체모집단에서 성공의 비율
        - $\pi = \frac{X}{N}$
            - N = 모집단의 수, X = 총 성공횟수
    - 표본에서의 성공비율
        - $p = \frac{X}{N}$
            - n = 표본의 크기, X = 표본에서의 성공횟수
    - 비율의 표집분포의 평균과 표준편차
        - 성공횟수의 기댓값과 분산
            - $E(X) = n\pi$
            - $Var(X) = n\pi(1-\pi)$
        - 비율의 표집분포의 기댓값과 분산
            - $E(p) = \pi$
            - $\sigma^{2}_{p} = \frac{\pi(1-\pi)}{n}$
- 이항분포의 정규근사법
    
    : 이항분포에서 $n$을 크게하여 정규분포로 간주하여 이항분포의 확률을 구하는 방법
    
    - 성공횟수 $X$에 해당하는 $Z$값
        - $Z = \frac{X_i - E(X)}{\sigma} = \frac{X_i - np}{\sqrt{npq}}$
    - 표본비율 $p$에 해당하는 $Z$값
        - $Z = \frac{p - \pi}{\sqrt{\pi(1-\pi)/n}{}}$
